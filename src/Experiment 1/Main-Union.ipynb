```python
#Load data from a csv
import csv
import os
import sys


def load_data(filename):
    data = []
    with open(filename , 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            data.append(row)

    return data

def save_data(filename, data):
    with open(filename, 'w') as f:
        writer = csv.writer(f)
        for row in data:
            writer.writerow(row)
```

```python
import pandas as pd

def build_data_from_csv(filename):
    df = pd.read_csv(filename)
    #Create input, labels pairs where input is all columns except the first and output is 0 for first and last 5000 rows and 1 for the rest
    X = df.iloc[:,1:]
    y = [0]*5000 + [1]*(df.shape[0]-10000) + [0]*5000

    return X, y
```

```python
# build data for all files in dataset folder. Use first 8 files for training and last 2 for testing

X_train = pd.DataFrame()
y_train = []
count = 0
for file in os.listdir('../dataset/part 1(10 eps)-unfiltered,no ref ch, no G,F,I ch'):
    print(file)
    if file.endswith('.csv'):
        count += 1
        if count > 8:
            break
        X, y = build_data_from_csv('../dataset/part 1(10 eps)-unfiltered,no ref ch, no G,F,I ch/' + file)
        X_train = pd.concat([X_train, X])
        y_train += y

X_test = pd.DataFrame()
y_test = []
count = 0
for file in os.listdir('../dataset/part 1(10 eps)-unfiltered,no ref ch, no G,F,I ch'):
    if file.endswith('.csv'):
        count += 1
        if count <= 8:
            continue
        X, y = build_data_from_csv('../dataset/part 1(10 eps)-unfiltered,no ref ch, no G,F,I ch/' + file)
        X_test = pd.concat([X_test, X])
        y_test += y




        
```

```python
#Fill all NaN with 0
X_train = X_train.fillna(0)
X_test = X_test.fillna(0)
```

```python
from sklearnex import patch_sklearn
patch_sklearn()
```

```python
# Make all columns in X_train and X_test equal by adding columns that are absent in one of them with all values 0
for col in X_train.columns:
    if col not in X_test.columns:
        X_test[col] = 0

for col in X_test.columns:
    if col not in X_train.columns:
        X_train[col] = 0
```

```python
len(X_train.columns)
```

```python
# Drop those columns from X_test that are not in X_train
X_test = X_test[X_train.columns]
```

```python


# Train a svm model
from sklearn.svm import SVC
model = SVC()
model.fit(X_train, y_train)

# Evaluate the model
model.score(X_test, y_test)

# Print accuracy, precision, recall, f1-score and confusion matrix
from sklearn.metrics import classification_report, confusion_matrix

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
# Try a different model
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)

model.score(X_test, y_test)

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
# Try a better model
from sklearn.ensemble import GradientBoostingClassifier

model = GradientBoostingClassifier()
model.fit(X_train, y_train)

model.score(X_test, y_test)

y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
