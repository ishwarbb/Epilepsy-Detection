```python
#Load data from a csv
import csv
import os
import sys

def load_data(filename):
    data = []
    with open(filename , 'r') as f:
        reader = csv.reader(f)
        for row in reader:
            data.append(row)

    return data

def save_data(filename, data):
    with open(filename, 'w') as f:
        writer = csv.writer(f)
        for row in data:
            writer.writerow(row)
```

```python
import pandas as pd

def build_data_from_csv(filename):
    df = pd.read_csv(filename)
    X = df.iloc[:,1:]  # Excluding the time access
    y = [0]*5000 + [1]*(df.shape[0]-10000) + [0]*5000

    return X, y
```

```python
X_all = pd.DataFrame()
y_all = []
for file in os.listdir('/kaggle/input/epilepsy-data/dataset/part 1(10 eps)-unfiltered,no ref ch, no G,F,I ch'):
    print(file)
    if file.endswith('.csv'):
        X, y = build_data_from_csv('/kaggle/input/epilepsy-data/dataset/part 1(10 eps)-unfiltered,no ref ch, no G,F,I ch/' + file)
        X_all = pd.concat([X_all, X])
        y_all += y
        
```

```python
import numpy as np
import pandas as pd

def build_data_from_csv(filename):
    df = pd.read_csv(filename)
    # Create input, labels pairs where input is all columns except the first and output is 0 for first and last 5000 rows and 1 for the rest
    X = df.iloc[:, 1:]
    y = [0] * 5000 + [1] * (df.shape[0] - 10000) + [0] * 5000

    # Splitting indices into train and test sets
    a1 = np.arange(0, 5000)
    a2 = np.arange(5000, df.shape[0] - 5000)
    a3 = np.arange(df.shape[0] - 5000, df.shape[0])

    t1 = list(np.random.choice(a1, int(len(a1) * 0.1), replace=False))
    tr1 = list(set(a1) - set(t1))
    t2 = list(np.random.choice(a2, int(len(a2) * 0.1), replace=False))
    tr2 = list(set(a2) - set(t2))
    t3 = list(np.random.choice(a3, int(len(a3) * 0.1), replace=False))
    tr3 = list(set(a3) - set(t3))

    t = t1 + t2 + t3
    tr = tr1 + tr2 + tr3

    # Splitting data and labels into train and test sets
    test_data = X.iloc[t]
    train_data = X.iloc[tr]
    test_labels = np.array(y)[t]
    train_labels = np.array(y)[tr]
    
    return train_data, test_data, list(train_labels), list(test_labels)
```

```python
X_train = pd.DataFrame()
X_test = pd.DataFrame()
y_train = []
y_test = []
for file in os.listdir('/kaggle/input/epilepsy-data/dataset/part 1(10 eps)-unfiltered,no ref ch, no G,F,I ch'):
    print(file)
    if file.endswith('.csv'):
        X1, X2, y1, y2 = build_data_from_csv('/kaggle/input/epilepsy-data/dataset/part 1(10 eps)-unfiltered,no ref ch, no G,F,I ch/' + file)
        X_train = pd.concat([X_train, X1])
        X_test = pd.concat([X_test, X2])
        y_train += y1
        y_test += y2
        
```

```python
#Repacing all NaN values with 0
X_train = X_train.dropna(axis=1)
X_test = X_test.dropna(axis=1)
```

```python
# Print accuracy, precision, recall, f1-score and confusion matrix
from sklearn.metrics import classification_report, confusion_matrix
```

```python
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier()
model.fit(X_train, y_train)

model.score(X_test, y_test)

```

```python
y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
# Try a different model
from sklearn.ensemble import RandomForestClassifier

print("Loading Model..")
model = RandomForestClassifier()
print("Fitting Model..")
model.fit(X_train, y_train)

model.score(X_test, y_test)


print("Evaluating Model...")
y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
# Train a svm model
from sklearn.svm import SVC
print("Loading Model..")
model = SVC()
print("Fitting Model..")
model.fit(X_train, y_train)

# Evaluate the model
model.score(X_test, y_test)

print("Evaluating Model..")
y_pred = model.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```
